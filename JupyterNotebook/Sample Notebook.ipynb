```python
benign_df = df.loc[df.diagnosis == 'B']
malignant_df = df.loc[df.diagnosis == 'M']

plt.scatter("smoothness_mean","perimeter_mean", data=malignant_df, color='pink', alpha=1)
plt.scatter("smoothness_mean","perimeter_mean", data=benign_df, color='#060', alpha=0.5)
plt.legend(['Malginant', 'Benign'])
plt.title('Smootheness Mean and Perimeter Mean of Cancer Cells')
plt.xlabel("Smoothness Mean")
plt.ylabel("Perimeter Mean")

plt.plot()
```
