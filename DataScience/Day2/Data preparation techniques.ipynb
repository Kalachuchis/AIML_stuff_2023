# Data Imputation
Data imputation is the process of filling in missing or null values in a dataset. This is often necessary because many machine learning algorithms and statistical models cannot handle missing data, and the performance of these models can be severely impacted if the missing data is not handled properly.

- Unordered Data
    - Mean imputation: This involves replacing missing values with the mean of the non-missing values for that variable.
    - Median imputation: This involves replacing missing values with the median of the non-missing values for that variable.
    - Mode imputation: This involves replacing missing values with the mode of the non-missing values for that variable.
    - Constant imputation : This involves replacing missing values with a specific constant value.
    - Random sample imputation : This involves replacing missing values with a random sample from the non-missing values of the variable.
- Ordered Data
    - Regression imputation : This involves replacing missing values with predicted values obtained through a regression model.
    - Forward - backward fill
 

## Unordered Data

### Measures of central tendency

```python
null_fare_df = df[df['Fare'].isna()].head()
null_fare_df
```

```python
#One way to impute the missing fares would be just to get the mean fare
fare_mean = df['Fare'].mean()

#Create a copy of dataframe 
df_mean_all_df = df.copy()

#Fill the dataframe with the mean
df_mean_all_df['Fare'] = df_mean_all_df['Fare'].fillna(fare_mean)
```

```python
df_mean_all_df.loc[null_fare_df.index]
```

```python
#But we could also get the different mean per pclass 
grouped = df.groupby(['Pclass'])['Fare'].mean()
grouped
```

```python
#And use those means to fill the null values
fare_mean = df['Pclass'].map(grouped)
fare_mean
```

```python
#Fill the null using the fare of Pclass
df_mean_per_class = df.copy()
df_mean_per_class['Fare'].fillna(fare_mean, inplace=True)
df_mean_per_class.loc[null_fare_df.index]
```

# Data Normalization
Data normalization: This is the process of scaling and centering the data, so that all variables are on the same scale, it could be done by different techniques such as:
- Min-Max normalization: Scale the data between 0 and 1
- Z-score normalization: Scale the data to have a mean of 0 and a standard deviation of 1

```python
#Let us see the min and max of two columns fare and age

print(f"Age min:{pd.to_numeric(df.Age).min()}")
print(f"Age max:{pd.to_numeric(df.Age).max()}")
print(f"Fare min:{pd.to_numeric(df.Fare).min()}")
print(f"Fare max:{pd.to_numeric(df.Fare).max()}")
```

```python
#If we do further analysis using these values one column will influence the result more due to its larger value
```

```python
df['Age'] = pd.to_numeric(df['Age'])
```

```python
import matplotlib.pyplot as plt
df_max_scaled = df[['Age','Fare']].copy()

df_max_scaled.plot.scatter(x = df_max_scaled.index,y = df_max_scaled.columns.tolist(),backend='plotly')
```

```python
#Min max scaling
df_min_max_scaled = df[['Age','Fare']].copy()
for col in df_min_max_scaled:
    df_min_max_scaled[col] = (df_min_max_scaled[col] - df_min_max_scaled[col].min()) / (df_min_max_scaled[col].max() - df_min_max_scaled[col].min())

#Use apply instead of loop
# df_min_max_scaled.apply(lambda x: (x-x.min())/(x.max()-x.min()))

df_min_max_scaled.plot.scatter(x = df_min_max_scaled.index,y = df_min_max_scaled.columns.tolist(),backend='plotly')
```

```python
#Zscore normalization
df_zscore_norm = df[['Age','Fare']].copy()
```

```python
#Pandas apply lambda is like a loop with axis = 0 as column-wise, and axis = 1 as row wise
df_zscore_norm = df_zscore_norm.apply(lambda x : (x - x.mean())/x.std(),axis=0)

df_zscore_norm.plot.scatter(x = df_min_max_scaled.index,y = df_min_max_scaled.columns.tolist(),backend='plotly')
```

Categorical data needs to be transformed because it cannot be used in its raw form by most machine learning models. Without transformation, the model would not be able to understand the categorical variables and would treat them as numeric variables, leading to inaccurate and unreliable results.

## Onehot encoding
One-hot encoding is a method of transforming categorical data into numerical data that can be understood by a machine learning model. The technique creates a new binary variable for each category in the feature, and assigns a value of 1 to the binary variable corresponding to the category that the sample belongs to, and a value of 0 to all the other binary variables.



```python
df = pd.get_dummies(df, columns=['Embarked'])
df.head()
```
