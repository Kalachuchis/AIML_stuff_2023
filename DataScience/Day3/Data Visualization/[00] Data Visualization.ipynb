# Line Graph

### Sample Problem: 
A Nutri world-wide firm wants to know how many people visits its website in a particular time. 
This analysis helps it control and monitor the website traffic.

This problem contains 2 variables: **Users** and **Time**.

This can be shown in 2 dimensional plot

```python
# website traffic data
# number of users/ visitors on the web site
web_customers = [123, 645, 950, 1290, 1630, 1450, 1034, 1295, 465, 205, 80]
# Time distribution (hourly)
time_in_hrs = list(range(7,18))
```

```python
# select the style of the plot
style.use('seaborn-pastel')
# plot the site traffic data (X-axis hrs and Y axis as number of users)
plt. plot(time_in_hrs, web_customers)
# set the title of the plot
plt. title('Website Traffic')
# set label for x axis
plt.xlabel('Hours')
# set label for y axis
plt.ylabel('Number of users')
```

## Controlling Line Patters and Colors

```python
'''
    You can change the line style of your graph using the linestyle param
'''
# plot the site traffic data (X-axis hrs and Y axis as number of users)
plt. plot(time_in_hrs, web_customers, color='blue', linestyle='--', linewidth=1.5)
# set the title of the plot
plt. title('Website Traffic')
# set label for x axis
plt.xlabel('Hours')
# set label for y axis
plt.ylabel('Number of users')
```

## Set Axis, Labels, and Legend Property

```python

plt. plot(time_in_hrs, web_customers, color='blue', label="web traffic", linewidth=1.5)
'''
    We can set the range of x and y axis
    by invoking the axis method.
    [x_min, x_max, y_min, y_max]
'''
plt.axis([6,17.5,0,2000])
plt. title('Website Traffic')
plt.xlabel('Hours')
plt.ylabel('Number of users')
plt.legend()
```

## Alpha and Annotation

Alpha is an attribute that controls the transparency of the line. The lower the alpha value, the more transparent the line is

```python
# We set another parameter (alpha)
plt. plot(time_in_hrs, web_customers, color='blue',
          label="web traffic", alpha=0.5)
plt.title('Website Traffic')
'''
    Annotate is used to annotate parts you want to highlight
    "Peak Visits" denotes the annotation text
    ha indicates the horizontal alignment - supported values are 'center', 'right', 'left'
    va indicates the vertical alignment - supported values are 'top', 'bottom', 'center', 'baseline', 'center_baseline'
    xytext indicates the text position
    xy indicates the arrow position
    arrowprops indicates the property of the arrow
'''
plt. annotate('Peak Visits', ha='center', va='top', xytext=(8, 1650),
                xy=(11,1630), arrowprops={'facecolor' : 'gray'})
plt.xlabel('Hours')
plt.ylabel('Number of users')
plt.legend()
```

## Subplots
A sublplot is used to display multiple plots in the same window. With subplot, you can arrange plots in a regular grid.

General syntax for subplot: **subplot(m,n,p)** where a plot is divided into **m rows** by **n cols** grid and creates an axis for a subplot in the position specified by **p**

### Layout and Spacing
You can adjust the horizontal and vertical spacing between subplots by using **subplots_adjust()** 

with **wspace** / **hspace** params for **width** and **height** space adjustment

![sample 2x1 plot](./subplot_2x1.jpg) !![sample 2 x 2 plot](./subplot_2x2.jpg)



```python
# Define random data
'''
    We can use np.random.uniform(start,stop,(rows,columns)) to generate numbers given a range
'''
temp_data = np.random.uniform(low=20, high=40, size=(24,))
wind_data = np.random.uniform(low=0, high=20, size=(24,))
humidity_data = np.random.uniform(low=40, high=80, size=(24,))
precipitation_data = np.random.uniform(low=0, high=80, size=(24,))
time_in_hrs = range(1,25)
```

```python
# Draw subplots
'''
    We can use plt.figure to first define the plot that we'll be using
    to draw our subplots

    figsize is (w,h) in inches
'''
plt.figure(figsize=(8,4))
plt.subplots_adjust(wspace=0.25)
'''
    This create a 1 x 2 plot 
    and defining the TEMP as the subplot in the first slot 
'''
plt.subplot(1,2,1)
plt.title("Temp")
plt.plot(time_in_hrs, temp_data, color='blue')
'''
    This create a 1 x 2 plot 
    and defining the WIND as the subplot in the second slot 
'''
plt.subplot(1, 2, 2)
plt.title("Wind")
plt.plot(time_in_hrs, wind_data, color='red')
```

```python
plt.figure(figsize=(6, 6))
plt.subplots_adjust(hspace=0.5)
'''
    This create a 2 x 1 plot 
    and defining the HUMIDITY as the subplot in the first slot 
'''
plt.subplot(2, 1, 1)
plt.title("Humidity")
plt.plot(time_in_hrs, humidity_data, color='blue')
'''
    This create a 2 x 1 plot 
    and defining the PRECIPITATION as the subplot in the second slot 
'''
plt.subplot(2, 1, 2)
plt.title("Precipitation")
plt.plot(time_in_hrs, precipitation_data, color='red')
```

```python
plt.figure(figsize=(9, 9))
plt.subplots_adjust(wspace=0.25)
plt.subplots_adjust(hspace=0.25)
'''
    This create a 2 x 2 plot 
    and defining the TEMP as the subplot in the 1st slot 
'''
plt.subplot(2, 2, 1)
plt.title("Temp")
plt.plot(time_in_hrs, temp_data, color='blue')
'''
    This create a 2 x 2 plot 
    and defining the WIND as the subplot in the 2nd slot 
'''
plt.subplot(2, 2, 2)
plt.title("Wind")
plt.plot(time_in_hrs, wind_data, color='red')
'''
    This create a 2 x 2 plot 
    and defining the HUMIDITY as the subplot in the 3rd slot 
'''
plt.subplot(2, 2, 3)
plt.title("Humidity")
plt.plot(time_in_hrs, humidity_data, color='green')
'''
    This create a 2 x 2 plot 
    and defining the PRECIPITATION as the subplot in the 4th slot 
'''
plt.subplot(2, 2, 4)
plt.title("Precipitation")
plt.plot(time_in_hrs, precipitation_data, color='magenta')
```

# Bar Chart

For this example, we'll use the breast cancer dataset.

```python
df =  pd.read_csv('breast-cancer-data.csv')
df.drop(columns='Unnamed: 32', inplace=True)
df.head()
```

```python
'''
    You can create a Bar Chart 
    using matplotlib natively

    You can use the colors parameters to assign color to each bar
'''

plt.bar(df.diagnosis.unique(), df.diagnosis.value_counts().sort_index(ascending=False), 
    color=['pink', 'green'], alpha=0.8)
plt.xlabel("Diagnosis Type")
plt.ylabel("Count")
plt.title("Breast Cancer Diagnosis Type Count")
print(df.diagnosis.unique())
print(df.diagnosis.value_counts().sort_index(ascending=False))
```

```python
'''
    You can also create a bar graph using pandas
    
    This is built on top of matplotlib 
    so you can still add your labels normally 
'''
df.diagnosis.value_counts().sort_values(ascending=True).plot(kind='bar')
plt.xlabel("Diagnosis Type")
plt.ylabel("Count")
plt.title("Breast Cancer Diagnosis Type Count")
```

## Stacked bar

```python
stacked_df = pd.read_csv("stacked_data.csv")
stacked_df
```

```python
'''
    We can use pandas pivot_table method to reshape the data
    
    We'll set the "branch" as the index
    We'll set the "sales_type" as the new column names
    And we'll set the values as the sales amount for each (branch, sales_type) cell
    margins = will add new col and row for grand_total
    agg_func = this function will be used when a grand total is created
'''
pivot = pd.pivot_table(data=stacked_df, index=['branch'], 
                        columns=['sales_type'], values='sales_amount', 
                       margins=True, aggfunc='sum', margins_name='sales_sum')
pivot.sort_values("sales_sum", ascending=False, inplace=True)
pivot.drop("sales_sum", inplace=True)
pivot.drop("sales_sum", inplace=True, axis=1)
pivot
```

```python
pivot.plot.bar(stacked=True, color=['green', 'tomato', "orange"], alpha=0.5)
plt.xlabel("SM Branch")
plt.ylabel("Annual Sales (in million Pesos)")
plt.title("Top Performing SM Stores")
```

```python
'''
    We can also use matplotlib to natively create a stacked bar
'''
years = range(2010,2016)
apple_prod = [35, 60, 90, 80, 65, 80]
orange_prod = [40, 80, 90, 70, 60, 80]

plt.bar(years, apple_prod, color='red', alpha=0.6, label='Apples')
plt.bar(years, orange_prod, bottom=apple_prod, color='orange', label='Oranges')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Count (in metric ton)")
plt.title("Philippine Fruit Production")
```

# Histogram

```python
'''
    We'll reuse the breast cancer data

    # of bins (in general) = sqrt(row count)
    range = Max - Min
    bin width = range / # of bins
'''
plt.hist(df.area_mean, bins=20, color='blue', alpha=0.5)
plt.title("Distribution of Cell Area Mean")
plt.xlabel("mean value")
plt.ylabel("count")
```

```python
df.area_mean
```

```python
'''
    We'll add a categorical variable to aid in visualizing
    the separation of distribution
'''
benign_df = df.loc[df.diagnosis == 'B']
malignant_df = df.loc[df.diagnosis == 'M']

plt.hist(malignant_df.area_mean, bins=20, color='red', alpha=0.5)
plt.hist(benign_df.area_mean, bins=20, color='blue', alpha=0.5)
plt.title("Distribution of Malignant vs Benign Cell Area Mean")
plt.ylabel("count")
plt.legend(['Malignant', 'Benign'])
```

```python
'''
    We can use pandas to generate a histogram
    column = column name to plot
    grid = bool to show or hide the grid lines
'''
df.hist(column='area_mean', bins=20, color='#86bf91',
        alpha=0.5, grid=False,)
plt.title("Distribution of Cell Area Mean")
plt.xlabel("mean value")
plt.ylabel("count")
```

```python
'''
        We can simultaneously plot based on an added categorical column
        In this example, we want to show the distribution of cell area mean
        based on their diagnosis (b or m)

        we added the "by" parameters and specified a categorical value
        Set the sharex to True if you want the scale of x to be equal
'''
df.hist(column='area_mean', by='diagnosis', bins=20, grid=False,
        figsize=(8, 10), layout=(2, 1), sharex=True, 
        color='#86bf91')
```

# Scatter Plot

```python
plt.scatter('smoothness_mean', "perimeter_mean",
            data=df, color='red', alpha=0.8)
plt.title("Smoothness Mean and Perimeter Mean of Cancer Cells")
plt.xlabel("Smoothness Mean")
plt.ylabel("Perimeter Mean")
```

```python
plt.scatter('smoothness_mean', "perimeter_mean",
            data=malignant_df, color='red', alpha=0.6)
plt.scatter('smoothness_mean', "perimeter_mean",
            data=benign_df, color='green', alpha=0.3)
plt.legend(['Malignant', 'Benign'])
plt.title("Smoothness Mean and Perimeter Mean of Cancer Cells")
plt.xlabel("Smoothness Mean")
plt.ylabel("Perimeter Mean")
```

```python
'''
    We can do the same thing with pandas
    but with additional step.

    The color (c) param accepts numeric data only
    We created a new column 'num_diagnosis' that uses
    the map function to create a numerical mapping
    of the categorical column

    you can see the list of available colormaps here: https://matplotlib.org/stable/tutorials/colors/colormaps.html
'''
df['num_diagnosis'] = df['diagnosis'].map(
    {
        "M": 1,
        "B": 0
    }
)
df.plot.scatter(x='smoothness_mean', y='perimeter_mean',
                c='num_diagnosis', colormap='winter')
plt.title("Smoothness Mean and Perimeter Mean of Cancer Cells")
plt.xlabel("Smoothness Mean")
plt.ylabel("Perimeter Mean")
```

# Heat maps

```python
rev_df = pd.read_csv('revenue_data.csv')
rev_df
```

```python
'''
    We'll create a pivot table to create a 
    dataframe for our heatmap

    Set sort to False to retain the order of row indices
'''
rev_pivot = pd.pivot_table(data=rev_df, index=['month'],
                       columns=['year'], values='revenue_amount', sort=False)
rev_pivot
```

```python
'''
    We need to get the col and index values for our legends later
'''
years = rev_pivot.columns.values
months = rev_pivot.index.values
print(months, years)
```

```python
'''
    Now, we'll use plt imshow to create a heatmap

    for the labels, the first pos. param. of _ticks
    is the list of int that the label it will replace.
    In this example, integer from 0 to number of labels

    cmaps is for colormap
'''
plt.figure(figsize=[5,10])
plt.imshow(rev_pivot, cmap='YlGn')
plt.xticks(range(len(years)), years)
plt.yticks(range(len(months)), months)
# plt.legend()
```

```python
'''
    In pandas, we can create a heatmap in the pivot table itself
'''

rev_pivot.style.background_gradient(cmap='YlGn')
```

# Pie Chart

```python
stacked_df
```

```python
# pie_pivot = pd.pivot_table(data=stacked_df, index=['sales_type'],
#                        columns=['branch'], values='sales_amount')
# pie_pivot
sm_north = stacked_df.loc[stacked_df['branch'] == 'North Ave.']
sm_north.sort_values(by='sales_amount', ascending=False, inplace=True)
sm_north
```

```python
'''
    When creating a pie chart, we need to sort the values (ascending)
    then plot the pie starting from 90 deg 
    and in clockwise direction

    you can add autopct param then specify the format
'''
plt.pie(sm_north.sales_amount, startangle=90,
        counterclock=False, autopct='%.2f%%')
plt.legend(sm_north['sales_type'].unique())
```

```python
'''
    You can also add a value to the "explode" param
    It separates the piece from the rest of the pie by x inch
'''
explode_list = [0.1] + [0] * (sm_north.shape[0] - 1)
plt.pie(sm_north.sales_amount, startangle=90,
        counterclock=False, autopct='%.2f%%',
        explode=explode_list)
plt.legend(sm_north['sales_type'].unique())
```

```python
'''
    We can create the same thing using pandas
'''
new_sm_north = sm_north.set_index('sales_type')
new_sm_north.plot.pie(y='sales_amount', autopct='%1.2f%%',
                      counterclock=False,  startangle=90, explode=explode_list)
plt.axis('off')
```
